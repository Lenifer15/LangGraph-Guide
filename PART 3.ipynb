## Conditional Edges

In this project, we will build a simple graph to demonstrate how conditional edges work in LangGraph. The goal is to create a minimal yet practical example that shows how decisions can influence the flow of execution within a graph. By using a simulated weather forecasting system, we can clearly observe how different outcomes trigger different paths in the graph, making it easier to understand the concept of conditional branching in a structured way.

The project uses a dummy weather prediction function that randomly forecasts either sunny or rainy weather, each with a 50% probability. Based on this result, the graph follows different edges, showcasing how LangGraph handles dynamic decision-making.

- Build a simple graph structure using LangGraph
- Implement a random-based weather prediction function
- Use conditional edges to route between “sunny” and “rainy” outcomes
- Understand how decision logic controls graph execution flow
- Learn the basics of conditional branching in a visual, practical way

### Step 2:

#### Define the nodes

The weather node takes a string as input and returns a pre-defined string.

```python
# Define a starting node. This node just returns a predefined string.
def weather(str):
  return "Hi! Well.. I have no idea... But... "
```

We define two more nodes, namely rainy_weather and sunny_weather. Remember nodes in langgraph are python functions or langchain runnables.

```python
# Define a node that returns rainy weather
def rainy_weather(str):
  return str + " Its going to rain today. Carry an umbrella."
```

```python
#Define a node that returns sunny weather
def sunny_weather(str):
  return str + " Its going to be sunny today. Wear sunscreen."
```

We will define another function forecast_weather() that will decide whether the weather is sunny or rainy.

```python
import random
from typing import Literal
```

```python
#create a forecast weather function that returns rainy or sunny weather based on a random number. 
def forecast_weather(str)->Literal["rainy", "sunny"]:
  if random.random() < 0.5:
    return "rainy"
  else:
    return "sunny"
```

### Step 3:

#### Create the graph instance

```python
from langgraph.graph import Graph, START, END
```

```python
# Create a new Graph
workflow = Graph()
```

### Step 4:

#### Add the nodes to the graph

```python
# Add the nodes
workflow.add_node("weather", weather)
workflow.add_node("rainy", rainy_weather)
workflow.add_node("sunny", sunny_weather)
```

### Step 5:

#### Add the edges

- The weather node is the entry point. So we will connect **START** to weather.
- The weather node decides where to go next based on the forecast_weather function.
- The nodes rainy and sunny are the terminal nodes. So they will be connected to the **END** node

```python
# Add the Edges
workflow.add_edge(START, "weather")
```

```python
workflow.add_conditional_edges("weather", forecast_weather)
workflow.add_edge("rainy", END)
workflow.add_edge("sunny", END)
```

### Step 6:

#### Compile the graph

```python
#Compile the graph
app = workflow.compile()
```

### Step 7:

#### Visualize the graph

```python
# Visualize your graph
from IPython.display import Image, display
png = app.get_graph().draw_mermaid_png()
```

```python
display(Image(png))
```

### Step 8:

#### Test the Graph

We test the graph by using app.invoke method.

```python
app.invoke('Hi! What does the weather look like? ')
```
