In this project we will learn how to call llm from LangGraph.

Calling llms such as ChatGPT, Gemini is easy. Let's look into it.

In this graph, there is only one node. When this graph is invoked, we will take a message as input and call an LLM. Then, the LLM responds back wit a response message.Then in return, the response message will be reverted back to the user.

#### Step 3:

Define the LLM Model

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model='gpt-4o')
```

#### Step 4:

**Define the LLM Node**

Define a function that will call the LLM thats defined in the previous block

```python
# define llm Node
def call_llm(message):
    return llm.invoke(message)
```

#### Create the Graph

We will create the graph that only has one node **call_llm**

The START and END nodes are special nodes defined in a LangGraph. The **Start** node represents the node that sends the user input to the graph. The **End** node represents the terminal node, after which the execution stops.

```python
from langgraph.graph import Graph, START, END

# Create a new Graph
workflow = Graph()
# Add the nodes
workflow.add_node("call_llm", call_llm)

# Add the Edges
workflow.add_edge(START, "call_llm")
workflow.add_edge("call_llm", END)

#Compile the workflow
app = workflow.compile()
```

#### Step 5:

Visualize the Graph using the **app.get_graph()** method. 

```python
# Visualize your graph
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

Lets test our graph. We send a message “hello” using **app.invoke(“hello”)**

```python
response = app.invoke("hello")
print(response.content)
```

### Conclusion

This project demonstrates that calling an LLM from LangGraph is straightforward even at its most minimal. A single-node graph — where user input enters, an LLM is invoked, and the response is returned — captures the essential pattern that all LangGraph applications build upon. The graph state carries messages in and out, the node is just a function that calls the model, and LangGraph handles the orchestration. Once you understand this foundation, scaling up to multi-node, multi-agent graphs is simply a matter of adding more nodes and defining how state flows between them.
