## StateGraph

State is a fundamental component of LangGraph that enables data to flow through the graph. Each node receives the current state as input, processes or modifies it, and then passes the updated state to the next node. This creates a continuous chain where every node builds on the output of the previous one, allowing the graph to evolve step by step until it reaches the terminal (END) node.

For example, you can include a property like count in the state schema to act as a counter. Each time the state passes through a node, this value can be incremented or updated, making it easy to track how many steps have been executed or how the data has changed throughout the graph.

So far we have built Graphs using the Graph instance. We can’t define a schema using Graph instance. In order to define a state schema, we wil have to use StateGraph instance. In this project, we will learn

- how to create a StateGraph instance.
- create a state schema
- update the schema

### Step 2:

#### Import required packages

```python
# Import equired packages
from langgraph.graph import StateGraph, START, END
from typing_extensions import TypedDict
```

### Step 3:

#### Define a state schema

We will use a schema of type TypedDict. Create a class called MyGraphState with two variables. count of datatype int and msg of datatype str.

```python
# Create the structure of the schema for the graph.
class MyGraphState(TypedDict):
  count: int
  msg: str
```

### Step 4:

#### Define nodes

We will define a function that will be used in multiple nodes in this graph.

```python
# Define your node
def counter(state: MyGraphState):
  state["count"] += 1
  state["msg"] = f"Counter function has been called {state['count']} time(s)"
  return state
```

We will add three nodes of the counter function

### Step 5:

#### Create a graph instance for the StateGraph

```python
# Create an instance of StateGraph with the structure of MyGraphState
workflow = StateGraph(MyGraphState)
```

### Step 6:

#### Add nodes to graph

```python
# Add three nodes to the workflow which are replicas of "counter"
workflow.add_node("Node1", counter)
workflow.add_node("Node2", counter)
workflow.add_node("Node3", counter)
```

### Step 7:

#### Add edges to the graph

```python
# Add the nodes one after another. 
# All the Nodes in the workflow take state as input, update the state and pass it to the next Node as input.
# Output of Node1 goes to Node2 as input. Output of Node2 goes to Node3 as input.
workflow.add_edge(START, "Node1")
workflow.add_edge("Node1", "Node2")
workflow.add_edge("Node2", "Node3")

workflow.add_edge("Node3", END)
```

### Step 8:

#### Compile the graph

```python
# Compile the workflow
app = workflow.compile()
```

### Step 9:

#### Visualize the graph

```python
# Visualize your graph
from IPython.display import Image, display
png = app.get_graph().draw_mermaid_png()

display(Image(png))
```

### Step 10:

#### Test your graph
